Remove numbers (points) from 'description'
  - how do we do this without removing other 'non-point' numbers? For example 'tremendous 100% varietal'... ' 'enjoy 2022-2030'
  - we could remove all 2-digit numbers. But there are likely mentions of 'this 50-50 blend or 70% cabernet 30% merlot, etc..
  - we could remove all 2-digit numbers in the range of points...
  - **OR for each row we could remove whatever that row's point score is**
    - turns out this removes many non-point numbers (ex: row 44 has 90 points and has '90% sangiovese' in the description)
  - Maybe we just take a naive approach and remove ALL numbers from the description. i don't mind doing this; the idea is 'can we prdecit price based on the language reviewers use to describe wines?', and although numbers like '90% sangiovese' or 'drink 2026-2030' contain information, they don't have much bearing on the *language* of the review per se

```python
# find all rows that contain the points value in description
match_index = {}
for row in range(len(df)):
  search_value = df.iloc[row]['points']
  if str(search_value) in str(df.iloc[row]['description']):
    match_index[row] = df.iloc[row]['description']
```
